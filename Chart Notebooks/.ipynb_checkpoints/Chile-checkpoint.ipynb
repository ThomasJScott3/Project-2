```python
# Do initial imports, read CSV, create dataframe and show first 5 values

import pandas as pd

df = pd.read_csv ('../Data/Chile/Chile CPI.csv')

df['Consumer Price Index (CPI)'] = df['CHLCPIALLMINMEI']

df.drop(columns=['CHLCPIALLMINMEI'], inplace=True)

df.head()
```

```python
# View last values

df.tail()
```

```python
# Import CSV, change column name and show top values of dataframe

import pandas as pd

df1 = pd.read_csv ('../Data/Chile/Chile Exchange Rate.csv')

df1['Exchange Rate (to USD)'] = df1['FXRATECLA618NUPN']

df1.drop(columns=['FXRATECLA618NUPN'], inplace=True)

df1.head()
```

```python
# View final values in the dataframe

df1.tail()
```

```python
# Create a new dataframe, change the column name and display the first value

import pandas as pd

df2 = pd.read_csv ('../Data/Chile/Chile GDP Percent Change.csv')

df2['GDP Percent Change'] = df2['CHLNGDPRPCH']

df2.drop(columns=['CHLNGDPRPCH'], inplace=True)

df2.head()
```

```python
# Display final values in the dataframe

df2.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
# Display scatter plot of data before the regression model is run

plt.scatter(df.index , df['Consumer Price Index (CPI)'])
```

```python
# Show scatter plot of second dataframe sans the regression model

plt.scatter(df1.index , df1['Exchange Rate (to USD)'])
```

```python
# Show scatter plot of third dataframe sans the regression model

plt.scatter(df2.index , df2['GDP Percent Change'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initialize the linear regression model

model = LinearRegression()
model
```

```python
# Reshape the dataframe

X = df.index.values.reshape(-1,1)
y = df['Consumer Price Index (CPI)'].values.reshape(-1,1)
```

```python
# Fit the model

model.fit(X, y)
```

```python
# Predict the 'y' values

predicted_y_values = model.predict(X)
predicted_y_values
```

```python
# Plot the linear regression line as a scatter plot

plt.scatter(X, y)

plt.title('Chile CPI Projection')

plt.plot(X, predicted_y_values, color='red')
```
